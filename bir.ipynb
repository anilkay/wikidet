```python
def getText(nodelist):
    # Iterate all Nodes aggregate TEXT_NODE
    rc = []
    for node in nodelist:
        if node.nodeType == node.TEXT_NODE:
            rc.append(node.data)
        else:
            # Recursive
            rc.append(getText(node.childNodes))
    return ''.join(rc)
```

```python
hamtext=getText(articles[2000:4000])
```

```python
hamtext[0:100]
```

```python
print("Çalışma tamamlandı.")
corpus=[]
```

```python
hamtext=hamtext.lower()
t_list = hamtext.split('\n')
for liste in t_list:
    corpus.append(liste.split())
```

```python
import numpy as np
from gensim.models import Word2Vec
```

```python
model=Word2Vec(corpus, size=100, window=5, min_count=2, sg=1)
```

```python
model["yapmak"]
```

```python
model.most_similar("amiga")
```

```python
model.save("wikimodel.model")
```
